```python
numeric_df = df[['Level',
                 'Description',
                 'SeriesRefSNDQ',
                 'Quarter',
                 'Weight',
                 'Amount']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Faqat numeric ustunlar
numeric_only = numeric_df.select_dtypes(include='number')

# Correlation
corr_matrix = numeric_only.corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, cmap='viridis', annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
level_mapping = {
    'Industry groups': 0
}
```

```python
description_mapping = {
    'Agriculture': 0,
    'Forestry and logging': 1,
    'Fishing, aquaculture, and agriculture, forestry and fishing support services': 2,
    'Mining': 3,
    'Food, beverage, and tobacco manufacturing': 4,
    'Textile, leather, clothing, and footwear manufacturing': 5,
    'Wood and paper products manufacturing': 6,
    'Printing': 7,
    'Petroleum, chemical, polymer, and rubber product manufacturing': 8,
    'Non-metallic mineral product manufacturing': 9
}
```

```python
df['Level'] = df['Level'].map(level_mapping)
df['Description'] = df['Description'].map(description_mapping)
```

```python
df.to_csv("updated_8.csv", index=False)
```
